```python
import pandas as pd
imperialism_loc = "/home/will/Documents/Projects/frc_counties/data/imperialism_map_start.csv"
```

```python
# read in csv
imp = pd.read_csv(imperialism_loc)
imp.head()
```

```python
# count each team
counts = {}
for index, row in imp.iterrows():
    teams = row["Teams"].split(",")
    for i in range(len(teams)):
        teams[i] = int(teams[i])
        if teams[i] not in counts:
            counts[teams[i]] = 1
        else:
            counts[teams[i]] += 1
```

```python
# print out top
sparsestTeams = [(k, counts[k]) for k in sorted(counts, key=counts.get, reverse=True)]
rank = 1
for team_tuple in sparsestTeams:
    print(str(rank) + ". " + str(team_tuple[0]) + " - " + str(team_tuple[1]) + " counties.")
    rank += 1
```
