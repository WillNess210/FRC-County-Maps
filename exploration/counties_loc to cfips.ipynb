```python
import pandas as pd
file_loc = "/home/will/Documents/Projects/frc_counties/data/counties_loc.csv"
```

```python
# read file
counties_loc = pd.read_csv(file_loc)
counties_loc = counties_loc.astype({"FIPS": str})
counties_loc.head()
```

```python
# update df with correct values
for idx, row in counties_loc.iterrows():
    nextval = str(row["FIPS"])
    if len(nextval) < 5:
        nextval = ("0" * (5 - len(nextval))) + nextval
    nextval = "c" + nextval
    counties_loc.at[idx, "FIPS"] = str(nextval)
counties_loc.head()
```

```python
# re-write over original
counties_loc.to_csv(file_loc, index=False)
```
